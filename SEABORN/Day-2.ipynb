```python
import matplotlib.pyplot as plt
import seaborn as sns

# Default white style set karo
plt.style.use("default")

# RC params update karo
plt.rcParams.update({
    "figure.facecolor": "white",  # figure ka background
    "axes.facecolor": "white",    # axis area ka background
    "axes.edgecolor": "black",    # axis border
    "axes.labelcolor": "black",   # labels ka rang
    "xtick.color": "black",       # x ticks color
    "ytick.color": "black"        # y ticks color
})

# Example plot
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip")

plt.show()
```

```python
# strip plot
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill')
```

```python
# jitter
sns.catplot(data=tips, x='day',y='total_bill',kind='strip',jitter=0.2,hue='sex')
```

```python
# swarmplot
sns.catplot(data=tips, x='day',y='total_bill',kind='swarm')
```
