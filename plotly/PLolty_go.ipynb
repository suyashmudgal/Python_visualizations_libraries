```python
import numpy as np
import pandas as pd   
import plotly.offline as pyo 
import plotly.graph_objs as go  
```

```python
match = pd.read_csv("matches.csv")
delivery = pd.read_csv("deliveries.csv")

ipl =delivery.merge(match,left_on = 'match_id',right_on ='id')
ipl.head()
```

```python
top50 = ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False ).head(50).index.tolist()
new_ipl = ipl[ipl['batsman'].isin(top50)]
```

```python
run = new_ipl.groupby('batsman')['batsman_runs'].sum()
balls= new_ipl.groupby('batsman')['batsman_runs'].count()

sr = (run/balls)*100

sr =sr.reset_index()
sr
```

```python
# Filter rows where player_dismissed is in top50
out = new_ipl[ipl['player_dismissed'].isin(top50)]
# Count number of times each player in top50 got out
nout = out['player_dismissed'].value_counts()
# Calculate average runs per dismissal for each batsman
avg = run / nout
avg = avg.reset_index()
avg.columns = ['batsman', 'avg']
# Merge with strike rate dataframe
avg = avg.merge(sr, on='batsman')
```

```python
temp_df = gap[gap['year'] == 2007]
temp_df
```

```python
import plotly.graph_objs as go

trace = go.Scatter(
    x=temp_df['lifeExp'],
    y=temp_df['gdpPercap'],
    mode='markers'
)

data = [trace]

layout = go.Layout(
    title="Life exp Vs GDP for 2007",
    xaxis={'title': "Life Exp"},   # no space here
    yaxis={'title': "GDP"}
)

fig = go.Figure(data=data, layout=layout)
fig.show()
```
