```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("stock.csv")   # <-- apni file ka naam
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.style.use('default')   # resets to white background
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['Close'])
plt.title("Stock Price Time Series")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decompose = seasonal_decompose(df['Close'], model='additive', period=30)
decompose.plot()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(df['Close'], lags=40)
plt.show()

plot_pacf(df['Close'], lags=40)
plt.show()
```

```python
train = df['Close'][:-30]
test = df['Close'][-30:]
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train, order=(1,1,1))
model_fit = model.fit()

forecast = model_fit.forecast(steps=30)
```

```python
plt.figure(figsize=(10,4))
plt.plot(train, label="Train")
plt.plot(test, label="Test")
plt.plot(forecast, label="ARIMA Forecast")
plt.legend()
plt.title("ARIMA Time Series Forecast")
plt.show()
```
